## Convolution Simple Example

```python
a = np.array([[1, 2, 0, 0],
            [5, 3, 0, 4],
             [0, 0, 0, 7],
            [9, 3, 0, 0]])
```

```python
plt.imshow(a)
```

```python
k = np.array([[1,1,1],
              [1,1,0],
              [1,0,0]])
```

```python
plt.imshow(k)
```

```python
from scipy import ndimage
```

```python
a_convolve = ndimage.convolve(a, k, mode='constant', cval=0)
```

```python
print(a_convolve)
```

```python
plt.imshow(a_convolve)
```

## Lets take a look at cat image

```python
from skimage import io
```

```python
from skimage.color import rgb2gray
```

```python
cat1 = io.imread("cat1.jpg")
```

```python
plt.imshow(cat1)
```

```python
cat1_gray =  rgb2gray(cat1)
```

```python
plt.imshow(cat1_gray)
```

```python
k
```

```python
cat1_convolve = ndimage.convolve(cat1_gray, k, mode='constant', cval=1.0)
```

```python
plt.imshow(cat1_convolve)
```

```python
laplacian = np.array((
	[0, 1, 0],
	[1, -4, 1],
	[0, 1, 0]), dtype="int")
# construct the Sobel x-axis kernel
sobelX = np.array((
	[-1, 0, 1],
	[-2, 0, 2],
	[-1, 0, 1]), dtype="int")
# construct the Sobel y-axis kernel
sobelY = np.array((
	[-1, -2, -1],
	[0, 0, 0],
	[1, 2, 1]), dtype="int")
```

```python
cat1_laplacian = ndimage.convolve(cat1_gray, laplacian, mode='constant', cval=1.0)
cat1_sx = ndimage.convolve(cat1_gray, sobelX, mode='constant', cval=1.0)
cat1_sy = ndimage.convolve(cat1_gray, sobelY, mode='constant', cval=1.0)
```

```python
plt.imshow(cat1_laplacian)
```

```python
plt.imshow(cat1_sx)
```

```python
plt.imshow(cat1_sy)
```
