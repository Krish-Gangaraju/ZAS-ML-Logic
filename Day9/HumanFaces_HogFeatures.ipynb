```python
from skimage import io
```

```python
from skimage.transform import rescale, resize, downscale_local_mean
```

```python
from matplotlib import pyplot as plt
```

```python
from skimage.feature import hog
from skimage import data, exposure
```

```python
p1 = io.imread("portrait1.jpg")
p2 = io.imread("portrait2.jpg")
p3 = io.imread("portrait3.jpg")
p4 = io.imread("portrait4.jpg")
```

```python
fig, axs = plt.subplots(2, 2,figsize=(8,8))
axs[0, 0].imshow(p1)
axs[0, 1].imshow(p2)
axs[1, 0].imshow(p3)
axs[1, 1].imshow(p4)
```

```python
p1_resized = resize(p1, (150, 150),
                       anti_aliasing=True)
p2_resized = resize(p2, (150, 150),
                       anti_aliasing=True)
p3_resized = resize(p3, (150, 150),
                       anti_aliasing=True)
p4_resized = resize(p4, (150, 150),
                       anti_aliasing=True)
```

```python
fig, axs = plt.subplots(2, 2,figsize=(8,8))
axs[0, 0].imshow(p1_resized)
axs[0, 1].imshow(p2_resized)
axs[1, 0].imshow(p3_resized)
axs[1, 1].imshow(p4_resized)
```

```python
fd_1, p1_hog_image = hog(p1_resized, orientations=8, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualize=True, multichannel=True)
fd_2, p2_hog_image = hog(p2_resized, orientations=8, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualize=True, multichannel=True)
fd_2, p3_hog_image = hog(p3_resized, orientations=8, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualize=True, multichannel=True)
fd_4, p4_hog_image = hog(p4_resized, orientations=8, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualize=True, multichannel=True)
```

```python
fig, axs = plt.subplots(2, 2,figsize=(8,8))
axs[0, 0].imshow(p1_hog_image)
axs[0, 1].imshow(p2_hog_image)
axs[1, 0].imshow(p3_hog_image)
axs[1, 1].imshow(p4_hog_image)
```
